```python
#Read the file in DataFrame
import pandas as pd
pd.read_csv('cricket.csv')
crick = pd.read_csv('cricket.csv')
crick
```

```python
#List the name of cricketer and their respective runs
crick [['name', 'runs']]
```

```python
# Find total wickets taken by them
crick['wickets'].sum()
```

```python
# Find average of catches taken
crick['catches'].mean()
```

```python
# Find the name of wicketkeeper
wicket_keeper = crick.loc[crick['stumpings']].max()
print ('The wicket keeper of the team is: ', wicket_keeper['name'])
```

```python
#print the name of bowler who played highest number of matches
crick[crick['wickets'] > 0].sort_values(by='matches',ascending=False)['name'].iloc[0]
```

```python
# Find average of all the bowlers
av = crick['runs']/crick['matches']
av
crick_mod = crick
crick_mod['av_run'] = av
crick_mod
```

```python
# Find name of the bowler with least bowling average (wickets per match)
av_min = crick['av'].min()
crick[crick['av']<= av_min]['name']
```

```python
# Print information about all players whose catches per match ratio is greater than 0.5
# store it in a csv file
av_catch = crick['catches']/crick['matches']
crick_mod = crick
crick_mod['av_catch'] = av_catch
crick_mod
crick_catch = crick_mod[crick_mod['av_catch']>0.5]
crick_catch
crick_catch.to_csv('out.csv')
```
